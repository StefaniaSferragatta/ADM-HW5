```python
def text_fix(names):
    
    for i in range(len(names)):
        
        names[i]=re.sub('\n','',names[i])
        
        names[i]=re.sub(f'^{str(i)}','',names[i]).lstrip()
        
    return names
```

```python
page_names=text_fix(names)
```
